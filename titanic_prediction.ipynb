```python
from sklearn.linear_model import LinearRegression
X=titanic[['Pclass','Age']]
```

```python
titanic['Sex'].replace(1, 'Female',inplace=True)
titanic['Sex'].replace(0, 'Male',inplace=True)
titanic.head()
```

```python
l=LinearRegression()
l.fit(X,titanic['Survived'])
```

```python
res.head()
```

```python
res["Age"].fillna(res.Age.mean(),inplace=True)
Z=res[['Pclass','Age']]
Y=l.predict(Z)
print(Y)
```

```python
y=np.where(Y>0.5,1,0)
res['Survived']=y
res.tail()
```

```python
test=res[['PassengerId','Survived']]
print(test)
```
