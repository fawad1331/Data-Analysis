```python
import numpy as np
import pandas as pd
```

```python
census_data=pd.read_csv(r"C:\Users\frali\Desktop\Data_analysis\censusindia2011.csv")
```

```python
census_data.head()
```

```python
census_total=census_data.loc[census_data['TRU']=='Total']
census_total
```

```python
census_total=census_total.loc[census_total['Level']=='STATE']
census_total
```

```python
census_states=census_total.sort_values('TOT_P')
census_states
```

```python
census_states.plot.barh('Name','TOT_P',figsize=(10,10))
```

```python
literacy_states=[]
for x in census_states['P_LIT']:
    literacy_states=census_states['P_LIT']/census_states['TOT_P']*100
census_states['LIT_RATES']=literacy_states
```

```python
census_states.plot.barh('Name','LIT_RATES',figsize=(10,8))
```

```python
lit_rate_f=[]
for x in census_states['P_LIT']:
    lit_rate=census_states['F_LIT']/census_states['TOT_P']*100
census_states['LIT_RATE_F']=lit_rate
```

```python
census_states=census_states.sort_values('LIT_RATE_F')
census_states.plot.barh('Name','LIT_RATE_F',figsize=(10,8))
```

```python
ilit=[]
for x in census_states['P_ILL']:
    ilit=census_states['P_ILL']/census_states['TOT_P']*100
census_states['ILLIT_RATE']=ilit
```

```python
census_states=census_states.sort_values('ILLIT_RATE')
census_states.plot.barh('Name','ILLIT_RATE',figsize=(10,10))
```
